Este notebook implementara un **wordcloud** con todas las palabras del dataset escogido para mostrar visualmente que palabras son las mas destacadas en nuestro conjunto de datos original y nuestro conjunto de datos lematizado. \
Para ello, necesitaremos dos datasets que genera la aplicacion **email-taxonomy**: el dataset original procesado (df_base) y dataset lematizado (df_base_lemmatized)

```python
import pickle
import random
from wordcloud import WordCloud
```

```python
def random_color_func(word=None, font_size=None, position=None,  orientation=None, font_path=None, random_state=None):
    list_colors = ['rgb(236,0,0)', 'rgb(230,0,0)', 'rgb(204,0,0)', 'rgb(179,0,0)', 'rgb(153,0,0)', \
                   'rgb(128,0,0)', 'rgb(0,0,0)', 'rgb(25,25,25)', 'rgb(51,51,51)', 'rgb(68,68,68)', \
                   'rgb(91,90,91)', 'rgb(109,109,109)', 'rgb(153,153,153)', 'rgb(194,194,194)']
    return random.choice(list_colors)
```

### Creamos un wordclod con los contenidos de los correos originales

```python
df = pickle.load(open("../tmp/df_base.p", "rb"))
df.head()
```

```python
# Join the different processed titles together.
long_string = ','.join(list(df['body'].head(500).values))
# Create a WordCloud object
wordcloud = WordCloud(background_color="white", max_words=5000, contour_width=4, color_func = random_color_func)
# Generate a word cloud
wordcloud.generate(long_string)
# Visualize the word cloud
wordcloud.to_image()
```

### Creamos un wordclod con los contenidos de los correos lematizados

```python
df_lemmatized = pickle.load(open("../tmp/df_base_lemmatized.p", "rb"))
df_lemmatized.head()
```

```python
# Join the different processed titles together.
long_string = ','.join(list(df_lemmatized['body_lemma'].head(500).values))
# Create a WordCloud object
wordcloud = WordCloud(background_color="white", max_words=5000, contour_width=3, contour_color='steelblue')
# Generate a word cloud
wordcloud.generate(long_string)
# Visualize the word cloud
wordcloud.to_image()
```
